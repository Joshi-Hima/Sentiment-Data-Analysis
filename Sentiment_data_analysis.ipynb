```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/TechAxis(Data Science)/Sentiment.csv')
df
```

```python
df.info()
```

```python
df['candidate_confidence'].value_counts()
```

```python
# lets see the id of the person who as the max retweet count

max_retweet = df['retweet_count'].max()
df.loc[max_retweet, 'id']
```

```python
df['relevant_yn'].value_counts()
```

```python
x = df['relevant_yn_confidence'] > 0.5
y = df['candidate_confidence'] > 0.5
plt.figure(figsize = (10,6))
plt.plot(x, y, marker = 'o')
plt.title('Relevant Confidence vs. Candidate Confidence')
plt.xlabel('Relevant Confidence')
plt.ylabel('Candidate Confidence')
plt.show()
```

```python
plt.figure(figsize = (10,6))
df['sentiment'].value_counts().plot(kind = 'bar', color = 'lightcoral')
plt.title('Sentiment Distribution', fontsize = 20)
plt.xlabel('Sentiment', fontsize = 20, labelpad = 15)
plt.ylabel('Count', fontsize = 20, labelpad = 15)
plt.show()
```

```python
# candidates counts

plt.figure(figsize = (10,6))
df['candidate'].value_counts().plot(kind = 'bar', color = 'lightcoral')
plt.title('Candidate Counts', fontsize = 20)
plt.xlabel('Candidate', fontsize = 20, labelpad = 15)
plt.ylabel('Count', fontsize = 20, labelpad = 15)
plt.show()
```

```python
# Group by sentiment and count the number of tweets
sentiment_counts = df['sentiment'].value_counts()
plt.figure(figsize=(8,8))
sentiment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Sentiments in Tweets')
plt.show()
```

```python
# Plot the histogram
plt.figure(figsize=(10,6))
plt.hist(df['sentiment_confidence'], bins=20, edgecolor='black')
plt.title('Distribution of Sentiment Confidence', fontsize = 20)
plt.xlabel('Sentiment Confidence', fontsize = 20, labelpad = 15)
plt.ylabel('Frequency', fontsize = 20, labelpad = 15)
plt.show()
```

```python
# lets see the donald trump tweets

trump_tweets = df[df['candidate'] == 'Donald Trump']
trump_tweets
```

```python
trump_sentiment_counts = trump_tweets['sentiment'].value_counts()
trump_sentiment_counts
```

```python
plt.figure(figsize = (10,6))
trump_sentiment_counts.plot(kind = 'bar', color = {'red', 'blue', 'orange'})
plt.title('Trump Sentiment Counts', fontsize = 20)
plt.xlabel('Sentiment', fontsize = 20, labelpad = 15)
plt.ylabel('Count', fontsize = 20, labelpad = 15)
plt.show()
```

```python
# lets see the tweet location of 13862 id

df.loc[df['id'] == 32, 'tweet_location']
```

```python
(df['candidate'] == 'Donald Trump').value_counts()
```

```python
# average retweet count for Donald Trump

avg_trump_retweet = trump_tweets['retweet_count'].mean()
avg_trump_retweet
```

```python
# scatter plot for relevant and candidate confidence for donald trump

grouped_data = trump_tweets.groupby('relevant_yn_confidence')['candidate_confidence'].mean().reset_index()

# Plot the bar graph
plt.figure(figsize=(10,6))
plt.bar(grouped_data['relevant_yn_confidence'], grouped_data['candidate_confidence'], color='blue')
plt.title('Relevant Confidence vs Candidate Confidence for Donald Trump')
plt.xlabel('Relevant Confidence')
plt.ylabel('Candidate Confidence')
plt.show()
```

```python
# top 10 most retweet

top_10_retweet = df.sort_values(by = 'retweet_count', ascending = False).head(10)
top_10_retweet
```

```python
# Plot the top 10 most retweeted tweets
plt.figure(figsize=(12,10))
plt.barh(top_10_retweet['text'], top_10_retweet['retweet_count'], color='skyblue')
plt.title('Top 10 Most Retweeted Tweets', fontsize = 20)
plt.xlabel('Retweet Count', fontsize = 20, labelpad = 15)
plt.ylabel('Tweet Text', fontsize = 20, labelpad = 15)
plt.show()
```
